```python
amz = pd.read_csv(r"/kaggle/input/ecommerce/amz_com-ecommerce_sample.csv")
fp = pd.read_csv(r"/kaggle/input/ecommerce/flipkart_com-ecommerce_sample.csv")
```

```python
amz = amz[['product_name','retail_price','discounted_price']]
fp = fp[['product_name','retail_price','discounted_price']]
```

```python
fp.isnull().sum()
```

```python
fp = fp.dropna(axis = 0, how ='any')
```

```python
amz[amz['retail_price']<0]
amz[amz['retail_price'] < 0] = None
amz[amz['discounted_price'] == 0] = None
amz = amz.dropna(axis = 0, how ='any')
```

```python
amz['product_name'] = amz['product_name'].apply(lambda a:a.lower())
fp['product_name'] = fp['product_name'].apply(lambda a:a.lower())
fp['retail_price'] = fp['retail_price'].apply(lambda a : int(a))
```

```python
fp.reset_index(inplace = True,drop = True)
amz.reset_index(inplace = True,drop = True)
```

```python
from sentence_transformers import SentenceTransformer, util
import tensorflow
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')

amazon_prod_embs = model.encode(amz.product_name,convert_to_tensor=True)
flipkart_prod_embs = model.encode(fp.product_name,convert_to_tensor=True)

cosine_sim_scores = util.cos_sim(amazon_prod_embs,flipkart_prod_embs)
```

```python
similar_prod_indexes = []

for i in range(19922):
    max_score_idxs = tensorflow.math.top_k(cosine_sim_scores[i],k=5,sorted=False).indices
    for idx in max_score_idxs:
        if idx != i:
            similar_prod_indexes.append(idx)
            
            break
```

```python
similar_prod_indexes = [x.numpy() for x in similar_prod_indexes]
```

```python
amz.product_name[1827], fp.product_name[similar_prod_indexes[1827]]
```

```python
# Extracting the details such as retail price and discounted price for both Flipkart and Amazon similar products
flipkart_prod_data = []
amazon_prod_data = []

for idx, prod_name in enumerate(amz.product_name):
    flipkart_prod_data.append(fp.iloc[similar_prod_indexes[idx]])
    amazon_prod_data.append(amz.iloc[idx])
```

```python
flipkart_prod_data = pd.DataFrame(flipkart_prod_data)
flipkart_prod_data.columns = ['Product name in Flipkart','Retail Price in Flipkart','Discounted Price in Flipkart']
amazon_prod_data = pd.DataFrame(amazon_prod_data)
amazon_prod_data.columns = ['Product name in Amazon','Retail Price in Amazon','Discounted Price in Amazon']
```

```python
flipkart_prod_data.head()
```

```python
amazon_prod_data['Retail Price in Amazon'] = amazon_prod_data['Retail Price in Amazon'].astype(np.float64)
amazon_prod_data['Discounted Price in Amazon'] = amazon_prod_data['Discounted Price in Amazon'].astype(np.float64)
```

```python
amazon_prod_data.head()
```

```python
flipkart_prod_final_data = flipkart_prod_data.copy()
flipkart_prod_final_data = flipkart_prod_final_data.reset_index(drop=True)
flipkart_prod_final_data.head()
```

```python
final_prod_data = pd.concat([flipkart_prod_final_data,amazon_prod_data],axis=1)
final_prod_data.to_csv('final_result.csv')
```
